### Loading the Data

```python
import pandas as pd

reaction_df = pd.read_csv("Reactions.csv", na_values=[""])
content_df = pd.read_csv("Content.csv")
reaction_types_df = pd.read_csv("ReactionTypes.csv")
```

### Cleaning the content data

```python
content_df['Category'] = content_df['Category'].str.replace('"', '')
```

### Remove User ID and URL columns as they not necessary

```python
content_df.drop(columns=['User ID'], inplace=True)
```

```python
content_df.drop(columns=['URL'], inplace=True)
```

```python
content_df
```

### Renaming Column "Type" to "Content Type" Just to be specific

```python
content_df.rename(columns={'Type': 'Content Type'}, inplace=True)
```

```python
content_df
```

### Cleaning Reactions.csv

```python
# Remove Rows with Blanks

reaction_df.dropna(inplace=True)
```

```python
# Remove "User ID" Column

reaction_df.drop(columns=['User ID'], inplace=True)
```

```python
# Renaming Type to Reaction Type

reaction_df.rename(columns={'Type': 'Reaction Type'}, inplace=True)
```

```python
reaction_df
```

```python
# Renaming Type to Reaction Type in reation_types_df

reaction_types_df.rename(columns={'Type': 'Reaction Type'}, inplace=True)
```

```python
reaction_types_df
```

## Part 2

### 1. Merging the Data Sets

```python
# Merge Reactions with Content
merged_df = reaction_df.merge(content_df, on='Content ID', how='left')

# Merge the result with Reaction Types
final_df = merged_df.merge(reaction_types_df, on='Reaction Type', how='left')
```

```python
final_df
```

### 2. Calculating the Total Scores for Each Category

```python
# Calculate total scores by category
category_scores = final_df.groupby('Category')['Score'].sum().reset_index()

# Sort categories by total score in descending order
top_categories = category_scores.sort_values(by='Score', ascending=False)

# Display the top 5 categories
top_5_categories = top_categories.head(5)
print(top_5_categories)
```

### 3. Saving the Final Data Set

```python
final_df.to_csv("Final_Cleaned_Data.csv", index=False)
top_5_categories.to_csv("Top_5_Categories.csv", index=False)
```

## Plotting the data

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set a style for the plots
sns.set_style("whitegrid")
```

```python
# Plot top 5 categories
plt.figure(figsize=(10, 6))
sns.barplot(data=top_5_categories, x='Category', y='Score', palette='viridis')
plt.title('Top 5 Content Categories by Total Score', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Score', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Distribution plot for scores
plt.figure(figsize=(8, 5))
sns.histplot(final_df['Score'], bins=30, kde=True, color='blue')
plt.title('Distribution of Reaction Scores', fontsize=16)
plt.xlabel('Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
reaction_type_scores = final_df.groupby('Reaction Type')['Score'].sum().reset_index()

# Pie chart for reaction type scores
plt.figure(figsize=(8, 8))
plt.pie(reaction_type_scores['Score'], labels=reaction_type_scores['Reaction Type'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('coolwarm', len(reaction_type_scores)))
plt.title('Reaction Types by Total Score', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart
plt.show()
```
